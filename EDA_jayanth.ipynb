```python
print("C: Analyze Distribution of Heart Disease ")
```

```python
plt.figure(figsize=(2,2))
ht['HeartDiseaseorAttack'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.show()
```

```python
print("Bivariate Analysis")
```

```python
print("A. Comparing the distribution of HighBP, HighChol & BMI between individuals with and without heart disease.")
```

```python
categorical_vars = ['HighBP', 'HighChol']

for col in categorical_vars:
    plt.figure(figsize=(3,4))
    sns.countplot(x=col, hue='HeartDiseaseorAttack', data=ht)
    plt.title(f'{col} vs Heart Disease')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.legend(title='Heart Disease', labels=['No', 'Yes'])
    plt.show()

# BMI is continuous (numerical) variable while HighBP & HighChol are categorical so we used Boxplot for BMI
plt.figure(figsize=(3,4))
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=ht)
plt.title('BMI Distribution by Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('BMI')
plt.show()
```

```python
print("B: Scatter Plots for the Age, BMI, PhysHlth, MentHlth")
```

```python
continuous_vars = ['Age', 'BMI', 'PhysHlth', 'MentHlth']
plt.figure(figsize=(3,3))
corr = ht[continuous_vars].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrice of Continuous Variables')
plt.show()
```

```python
print("C: Analyze the prevalence of heart disease across Demographic groups like: age, gender & education using bar charts.")
```

```python
ht.groupby('AgeGroup')['HeartDiseaseorAttack'].mean().plot(kind='bar', color='skyblue')
plt.title('Heart Disease by Age Group')
plt.ylabel('Percentage')
plt.show()

ht.groupby('Sex')['HeartDiseaseorAttack'].mean().plot(kind='bar', color='lightcoral')
plt.title('Heart Disease by Gender')
plt.ylabel('Percentage')
plt.show()

ht.groupby('Education')['HeartDiseaseorAttack'].mean().plot(kind='bar', color='lightgreen')
plt.title('Heart Disease by Education Level')
plt.ylabel('Percentage')
plt.show()
```
