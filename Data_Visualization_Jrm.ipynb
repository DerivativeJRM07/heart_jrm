```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ht=pd.read_csv('heart.xls.csv')
```

***Ploting Bar charts***

a. Visualize Categorical Variables:

Plot **bar charts** to visualize the distribution of categorical variables such as Smoker, PhysActivity, Fruits and Veggies.


```python
plt.figure(figsize=(6,3))
ht['Smoker'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Smoker Distribution')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,3))
ht['PhysActivity'].value_counts().plot(kind='bar', color='red')
plt.title('PhysActivity Distribution')
plt.xlabel('PhysActivity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,3))
ht['Fruits'].value_counts().plot(kind='bar', color='orange')
plt.title('Fruits Distribution')
plt.xlabel('Fruits')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,3))
ht['Veggies'].value_counts().plot(kind='bar', color='Green')
plt.title('Veggies Distribution')
plt.xlabel('Veggies')
plt.ylabel('Count')
plt.show()
```

b. Compare Heart Disease Across Categories:

Use **grouped bar charts** to compare the prevalence of heart disease across different categorical variables.

```python
cols = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']

for col in cols:
    # Create a grouped table (cross-tab)
    grouped = pd.crosstab(ht[col], ht['HeartDiseaseorAttack'])

    grouped.plot(kind='bar', figsize=(6,4))
    plt.title(f'Heart Disease by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.legend(title='Heart Disease', labels=['No', 'Yes'])
    plt.show()
```

***Exploring other visualization techniques***

D. **Line Plots:**  Display trends over time or across ordered categories.


```python
age_groups = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
avg_bmi = [22.5, 24.1, 26.3, 27.8, 28.5, 27.9]

plt.plot(age_groups, avg_bmi, marker='o', linestyle='-', color='b')
plt.title('Average BMI Across Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Average BMI')
plt.grid(True)
plt.show()
```

E: **Violin Plots:** Combine aspects of box plots and density plots to show the distribution of data.

```python
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=ht)

plt.title('Distribution of BMI by Heart Disease Status')
plt.xlabel('Heart Disease (Yes/No)')
plt.ylabel('BMI')
plt.show()
```
