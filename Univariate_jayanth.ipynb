```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ht=pd.read_csv('heart.xls.csv')
```

```python
ht.head()
```

```python
print('A: Histogram Distributions of Individual Variables of Age, BMI, MentHlth & PhysHlth')
```

```python
plt.figure(figsize=(3,4))
sns.distplot(ht['Age'],kde=False);
```

```python
plt.figure(figsize=(3,4))
sns.distplot(ht['BMI'],kde=False);
```

```python
plt.figure(figsize=(5,4))
sns.distplot(ht['MentHlth'],kde=False);
```

```python
plt.figure(figsize=(5,4))
sns.distplot(ht['PhysHlth'],kde=False);
```

```python
print('B: Pie chart distribution prevalence of health conditions like: HeartDiseaseorAttack, HighBP, HighChol, Stroke, Diabetes, GenHlth, MentHlth, PhysHlth, DiffWalk ')
```
